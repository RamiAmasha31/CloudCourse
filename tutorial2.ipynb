```python
from google.colab import drive
from ipywidgets import widgets, Output
from IPython.display import display, clear_output
import json
drive.mount('/content/drive')

students_file_path = '/content/drive/MyDrive/students.json'
data = [
    {
        "fname": "Rami",
        "lastname": "Amasha",
        "braudeMail": "rami.amasha@e.braude.ac.il",
        "courses_list": ["data mining", "crypto", "infi"],
        "urls": "https://google.com",
        "fav_prog": None
    },
    {
        "fname": "Alice",
        "lastname": "Johnson",
        "braudeMail": "alice.johnson@e.braude.ac.il",
        "courses_list": ["Data structures", "crypto"],
        "urls": "https://example.com/alice_profile",
        "fav_prog": None
    },
    {
        "fname": "Bob",
        "lastname": "Smith",
        "braudeMail": "bob.smith@e.braude.ac.il",
        "courses_list": ["crypto"],
        "urls": "https://example.com/bob_profile",
        "fav_prog": None
    },
    {
        "fname": "Charlie",
        "lastname": "Brown",
        "braudeMail": "charlie.brown@e.braude.ac.il",
        "courses_list": ["algebra", "infi"],
        "urls": "https://example.com/charlie_profile",
        "fav_prog": None
    }
]
with open(students_file_path, 'w') as file:
    json.dump(data, file)
from google.colab import drive
from ipywidgets import widgets, Output
from IPython.display import display, clear_output

drive.mount('/content/drive')

students_file_path = '/content/drive/MyDrive/students.json'

with open(students_file_path, 'r') as file:
    students_data = json.load(file)

first_names = [student['fname'] for student in students_data]
last_names = [student['lastname'] for student in students_data]
emails = [student['braudeMail'] for student in students_data]
courses_list = [student['courses_list'] for student in students_data]
urls = [student['urls'] for student in students_data]
favorite_program_input_values = [''] * len(students_data)

dropdown = widgets.Dropdown(options=first_names, description='Select:')
output = Output()

favorite_program_input = widgets.Text(value='', description='Favorite Program:')

update_button = widgets.Button(description='Update Information')

# function to handle dropdown value changes
def on_dropdown_change(change):
    with output:
        clear_output(wait=True)  # Clear previous output
        selected_option = change['new']
        idx = first_names.index(selected_option)
        print(f'Selected Option: {selected_option}')
        print(f'Last Name: {last_names[idx]}')
        print(f'Email: {emails[idx]}')
        print(f'Courses: {courses_list[idx]}')
        print(f'Urls: {urls[idx]}')


def update_information(button):
    with output:
        clear_output(wait=True)  # Clear previous output
        selected_option = dropdown.value
        idx = first_names.index(selected_option)
        favorite_program = favorite_program_input.value

        favorite_program_input_values[idx] = favorite_program

        # Update the information in the loaded data
        students_data[idx]['fav_prog'] = favorite_program

        # Write the updated data back to the JSON file
        with open(students_file_path, 'w') as file:
            json.dump(students_data, file, indent=2)

# Attach the function to the button click event
update_button.on_click(update_information)

# Observe the value changes in the dropdown
dropdown.observe(on_dropdown_change, names='value')

# Display the widgets
display(dropdown, favorite_program_input, update_button, output)

# Initially trigger the on_dropdown_change function with the first student's name
on_dropdown_change({'new': first_names[0]})
```
